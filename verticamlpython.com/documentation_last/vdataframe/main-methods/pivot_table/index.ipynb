# vDataFrame.pivot_table

```python
vDataFrame.pivot_table(columns: list,
                       method: str = "count",
                       of: str = "",
                       max_cardinality: tuple = (6, 6),
                       h: tuple = (None, None),
                       show: bool = True,
                       cmap: str = '',
                       with_numbers: bool = True)
```

Draws a pivot table of one or two columns based on an aggregation.

### Parameters

<table id="parameters">
    <tr> <th>Name</th> <th>Type</th> <th>Optional</th> <th>Description</th> </tr>
    <tr> <td>columns</td> <td><div class="type">list</div></td> <td><div class = "no">&#10060;</div></td> <td>List of the vcolumns names. The list must have one or two elements.</td> </tr>
    <tr> <td><div class="param_name">method</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>The method used to aggregate the data.<br>
                                                    <ul>
                                                        <li><b>count :</b> Number of elements.</li>
                                                        <li><b>density :</b> Percentage of the distribution.</li>
                                                        <li><b>mean :</b> Average of the vcolumn 'of'.</li>
                                                        <li><b>min :</b> Minimum of the vcolumn 'of'.</li>
                                                        <li><b>max :</b> Maximum of the vcolumn 'of'.</li>
                                                        <li><b>sum :</b> Sum of the vcolumn 'of'.</li>
                                                        <li><b>q% :</b> q Quantile of the vcolumn 'of' (ex: 50% to get the median).</li>
                                                        </ul></td> </tr>
    <tr> <td><div class="param_name">of</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>The vcolumn used to compute the aggregation.</td> </tr>
    <tr> <td><div class="param_name">h</div></td> <td><div class="type">tuple</div></td> <td><div class = "yes">&#10003;</div></td> <td>Interval width of the vcolumns 1 and 2 bars.     <tr> <td><div class="param_name">cmap</div></td> <td><div class="type">str</div></td> <td><div class = "yes">&#10003;</div></td> <td>Color Map.</td> </tr>

</table>

### Returns

<b>vDataFrame</b> : self

### Example

```python
from vertica_ml_python.learn.datasets import load_titanic
titanic = load_titanic()
print(titanic)
```

```python
titanic.pivot_table(["survived", "pclass"])
```

```python
# Avg of survived partitioned by pclass, age
titanic.pivot_table(["age", "pclass"], method = "avg", of = "survived")
```

```python
# Avg of survived partitioned by age, fare
# age will be discretized using an interval of 10 years
# fare will be discretized using an interval of 50
titanic.pivot_table(["age", "fare"], 
                    method = "avg", 
                    of = "survived",
                    h = (10, 50))
```
