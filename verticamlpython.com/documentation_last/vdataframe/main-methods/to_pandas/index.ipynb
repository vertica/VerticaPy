# vDataFrame.to_pandas

```python
vDataFrame.to_pandas()
```

Converts the vDataFrame to a pandas DataFrame. Be careful as you'll move data in memory.

### Returns

<b>pandas.DataFrame</b> : The pandas.DataFrame of the current vDataFrame relation.

### Example

```python
from vertica_ml_python.learn.datasets import load_titanic
titanic = load_titanic()
print(titanic)
```

```python
titanic.to_pandas()
```
