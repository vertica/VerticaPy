```python
%%sql
DROP SCHEMA IF EXISTS telco CASCADE;
CREATE SCHEMA telco;
```

```python
# Data Ingestion - 1
campaign_telco = read_csv('campaign_telco.csv', 
                          table_name = "campaign", 
                          schema = "telco")
display(campaign_telco)
```

```python
# Data Ingestion - 2
customer_telco = read_csv('customer_telco.csv', 
                          table_name = "customer", 
                          schema = "telco")
display(customer_telco)
```

## Data Preparation

```python
# Marketing campaign
campaign_telco["conversion"] = "CASE WHEN ConversionChannel IS NOT NULL THEN 1 ELSE 0 END"

# Joining the datasets
customer_campaign = customer_telco.join(
             campaign_telco.search(conditions = "conversionchannel IS NOT NULL"),
             how = "left",
             on = {"customerID": "customerID"},
             expr1 = ["*"],
             expr2 = ["campaign", 
                      "channel", 
                      "open", 
                      "click", 
                      "saleschannel", 
                      "conversionchannel",
                      "offer",
                      "conversion",
                      "target"],)
customer_campaign["conversion"] = "CASE WHEN ConversionChannel IS NOT NULL THEN 1 ELSE 0 END"
display(customer_campaign)
```

## Data Exploration

```python
# DANGER Sales Channel is highly correlated with Conversion Channel
# BUT it is not causal
campaign_telco.corr_pvalue("SalesChannel", "ConversionChannel", method="cramer")
```

```python
# Churners will not convert
customer_campaign.corr_pvalue("churn", "conversion", method="biserial")
```

```python
# Conversions
customer_campaign.search("conversionchannel IS NOT NULL ").hchart(
                                                        x="channel",
                                                        y="target",
                                                        kind="bar",
                                                        drilldown=True)
```

```python
# target KPI
campaign_telco.groupby("target", 
                       expr = ["COUNT(*) AS volume", 
                               "SUM(open) AS open", 
                               "SUM(click) AS click",
                               "SUM(conversion) AS conversion", 
                               "SUM(open) / COUNT(*) AS open_rate",  
                               "SUM(click) / COUNT(*) AS click_rate", 
                               "SUM(conversion) / COUNT(*) AS conversion_rate"])
```

## Features Engineering

```python
channels = campaign_telco["channel"].distinct()
for c in channels:
    campaign_telco[c + "_click"] = """
        SUM(CASE 
                WHEN channel = '{}' AND click = 1 THEN 1 
                ELSE 0 
            END) OVER (PARTITION BY CustomerID)""".format(c)
    campaign_telco[c + "_click"].astype("bool")
    campaign_telco[c + "_open"]  = """
        SUM(CASE 
                WHEN channel = '{}' AND open  = 1 THEN 1 
                ELSE 0 
            END) OVER (PARTITION BY CustomerID)""".format(c)
    campaign_telco[c + "_open"].astype("bool")
    campaign_telco["channel_" + c] = """
        SUM(CASE 
                WHEN channel = '{}' THEN 1 
                ELSE 0 
            END) OVER (PARTITION BY CustomerID)""".format(c)
    campaign_telco["channel_" + c].astype("bool")
campaign_telco = campaign_telco.search("conversion = 1")
campaign_telco.drop(["conversion",
                     "open",
                     "click",])
```

```python
print(campaign_telco.current_relation())
```

## Machine learning 

```python
from verticapy.learn.ensemble import RandomForestClassifier

set_option("random_state", 1)

train, test = campaign_telco.train_test_split(random_state=1)
predictors  = campaign_telco.get_columns(exclude_columns = ['ConversionChannel', 
                                                           'CustomerID',]) 
model = RandomForestClassifier("telco.attribution_tesla_model",
                               n_estimators = 100, 
                               sample = 0.75, 
                               max_depth = 10, 
                               min_samples_leaf = 4, 
                               max_leaf_nodes= 3000,
                               nbins = 100)
model.fit(train, 
          predictors, 
          "ConversionChannel", 
          test)
model.classification_report()
```

```python
model.score(method="accuracy")
```

```python
model.confusion_matrix()
```
