```python
# Data Ingestion
football = read_csv('football_2021.csv', 
                    table_name = "games", 
                    schema = "football",
                    parse_n_lines = 1000,)
display(football)
```

### Games Transposition

```python
# Some team names changed through time
for team in ["home_team", "away_team"]:
    football[team].decode('German DR', 'Germany',
                          'Czechoslovakia', 'Czech Republic',
                          'Yugoslavia', 'Serbia',
                          football[team])
football.to_db("football.games_clean", relation_type = "table")   

# Each team can be at home (on the right) or away (on the left). 
# We intervert to compute the aggregations = Translation
football_1 = vDataFrame("football.games_clean")
football_1["duplicate_id"] = "1"
football_1["home_team"].rename("team1")
football_1["home_score"].rename("team1_score")
football_1["away_team"].rename("team2")
football_1["away_score"].rename("team2_score")
football_1["neutral"].decode(True, 0, 1)

football_2 = vDataFrame("football.games_clean")
football_2["duplicate_id"] = "2"
football_2["home_team"].rename("team2")
football_2["home_score"].rename("team2_score")
football_2["away_team"].rename("team1")
football_2["away_score"].rename("team1_score")
football_2["neutral"].decode(True, 0, 2)

# Merging the 2 datasets
all_games = football_1.append(football_2)
all_games["neutral"].rename("home_team_id")
display(all_games)
```

### Data Preparation

```python
# game's winner
all_games["winner"] = st.case_when(all_games["team1_score"] > all_games["team2_score"], 1, 0)

# is_home feature
all_games["home"] = """CASE 
                         WHEN team1 = country 
                             AND tournament NOT LIKE ('%qualification%') 
                             THEN 1 
                         ELSE 0 
                       END"""

# game's year
all_games["year"] = st.year(all_games["date"])

# Getting all the European teams
uefa_team_list = all_games.search(all_games["tournament"] == 'UEFA Euro qualification')["team1"].distinct()

# We filter by looking only at World Cup and Euro
all_games = all_games.search(conditions = football["tournament"]._in(['UEFA Euro qualification',
                                                                      'UEFA Euro',
                                                                      'FIFA World Cup qualification',
                                                                      'FIFA World Cup']))
# Identifier to know if a european team is playing
all_games["uefa"] = st.case_when(all_games["team1"]._in(uefa_team_list), 1, 0)

# Identifier to know the tournament type
all_games["type_tournament"] = """CASE 
                                    WHEN tournament LIKE ('%UEFA%') 
                                        THEN 'UEFA' 
                                    ELSE 'FIFA' 
                                   END"""
display(all_games)
```

```python
# We use sessionization to merge qualifications & competitions
all_games["date"].astype("datetime")
all_games.sessionize(ts = "date", 
                     by = ["tournament"], 
                     session_threshold = "1 year", 
                     name = "edition")

# The first World Cup didn't have qualifications, we increment the edition
all_games["edition"] = st.case_when(all_games["tournament"] == 'FIFA World Cup qualification',
                                    all_games["edition"] + 1, all_games["edition"])
display(all_games[["team1", "team2", "year", "tournament", "edition"]])
```

```python
# Competition Ranks based on the different games and rounds
all_games.analytic('row_number', 
                   order_by = {"date": "desc",}, 
                   by = ["type_tournament",
                         "edition",
                         "duplicate_id",], 
                   name = "tournament_round")
all_games["tournament_round"] = """
                        CASE WHEN type_tournament = 'FIFA' THEN
                            CASE WHEN tournament_round <= 2 THEN tournament_round
                                 WHEN tournament_round <=  4 THEN 3
                                 WHEN tournament_round <=  8 THEN 4
                                 WHEN tournament_round <=  16 THEN 5
                                 WHEN tournament = 'FIFA World Cup' THEN 6
                                 ELSE 7
                            END
                        ELSE
                            CASE WHEN tournament_round = 1 THEN 1
                                 WHEN tournament_round <=  3 THEN 2
                                 WHEN tournament_round <=  7 THEN 3
                                 WHEN tournament_round <=  15 THEN 4
                                 WHEN tournament = 'UEFA Euro' THEN 5
                                 ELSE 6
                            END  
                        END"""
all_games["tournament_round_points"] = """SUM((team1_score > team2_score):: int * 3 +
                                              (team1_score = team2_score):: int)
                                          OVER (PARTITION BY type_tournament,
                                                             edition,
                                                             tournament_round,
                                                             team1)"""
all_games["tournament_diff_goals"] = """SUM(team1_score - team2_score)
                                        OVER (PARTITION BY type_tournament,
                                                           edition,
                                                           tournament_round,
                                                           team1)"""
all_games["tournament_round_goals"] = """SUM(team1_score)
                                         OVER (PARTITION BY type_tournament,
                                                            edition,
                                                            tournament_round,
                                                            team1)"""

# We compute the rank of each team by competition
all_games.analytic('row_number', 
                   order_by = {"tournament_round": "asc",
                               "winner": "desc",
                               "tournament_round_points": "desc",
                               "tournament_diff_goals": "desc",
                               "tournament_round_goals": "desc",}, 
                   by = ["type_tournament",
                         "edition",], 
                   name = "order_tournament")

# We aggregate our data to obtain only teams with kpi
teams_kpi = all_games.groupby(["edition",
                               "type_tournament",
                               "team1 AS team",
                               'uefa'],
                              [st.min(all_games["order_tournament"])._as("order_tournament"),
                               st.max(all_games["home"])._as("home"),
                               st.max(all_games["year"])._as("year"),
                               st.count(all_games["tournament"])._as("number_games"),
                               """SUM(CASE WHEN ((type_tournament = 'UEFA' AND tournament_round >= 5)
                                             OR (type_tournament = 'FIFA' AND tournament_round >= 6))
                                             THEN (team1_score > team2_score):: int * 3 +
                                                  (team1_score = team2_score):: int
                                             ELSE 0
                                       END) AS total_points"""])

# Smoothing the rank because of duplicates                         
teams_kpi.analytic('row_number', 
                 order_by = {"order_tournament": "asc"}, 
                 by = ["type_tournament", 
                       "edition"], 
                 name = "rank")
teams_kpi["order_tournament"].drop()

# Cleaning the EURO 2021
teams_kpi["year"] = """CASE 
                         WHEN type_tournament='UEFA' AND edition = 15 THEN 2021
                         ELSE year 
                       END"""
for elem in ["rank", "total_points", "number_games",]:
    teams_kpi[elem] = """CASE 
                           WHEN type_tournament='UEFA' AND edition = 15 THEN NULL
                           ELSE {} 
                         END""".format(elem)

display(teams_kpi[["year", 
                   "type_tournament", 
                   "team", 
                   "rank"]].search("year = 2018").sort({"year": "desc",
                                                        "type_tournament": "asc",
                                                        "rank": "asc"}))
```

### Features Engineering

```python
# Cup Winners
teams_kpi["winner"] = st.case_when(teams_kpi["rank"] == 1, 1, 0)

# inputs are our KPIS
# predictors are our KPIs lags
# The response is the rank
inputs, predictors, response  = [], [], "rank"

# World Cup & UEFA must be seen as two independent competitions
for elem in ["number_games",
             "total_points",
             "rank",
             "winner"]:
    teams_kpi[f"{elem}_world_cup"] = st.case_when(teams_kpi["type_tournament"] == 'FIFA',
                                                  teams_kpi[elem], None)
    teams_kpi[f"{elem}_euro"] = st.case_when(teams_kpi["type_tournament"] == 'UEFA',
                                             teams_kpi[elem], None)
    inputs += [f"{elem}_world_cup", f"{elem}_euro"] 

# Lags of our features to avoid information bias
for elem in inputs: 
    teams_kpi.analytic(func = 'lag',
                       columns = [elem], 
                       order_by = {"year": "asc"}, 
                       by = ["team"] , 
                       offset = 1,
                       name = f"lag_{elem}")
    predictors += [f"lag_{elem}"]

# We impute missing values
for elem in predictors: 
    teams_kpi[elem].apply(f"CASE WHEN {elem} IS NULL THEN LAST_VALUE({elem} IGNORE NULLS) OVER(PARTITION BY team ORDER BY year ASC) ELSE {elem} END")
    teams_kpi[elem].fillna(method='mean', by=["team", "type_tournament"])
teams_kpi["rank"].fillna(method='mean', by=["team", "type_tournament"])  

# We compute moving windows of our predictors & target
predictors_tmp = []
for elem in predictors:
    teams_kpi.rolling(func = "avg",
                      window = (-3, 0),
                      columns = elem,
                      by = ["team"],
                      order_by = ["year"],
                      name = f"{elem}_1_3")
    predictors_tmp += [f"{elem}_1_3"]
predictors += predictors_tmp
teams_kpi.rolling(func = "avg",
                  window = (-3, -1),
                  columns = "rank",
                  by = ["team"],
                  order_by = ["year"],
                  name = "rank_1_3")
predictors += ["rank_1_3", "home"]
display(teams_kpi)
```

### Machine Learning

```python
from verticapy.learn.linear_model import LinearRegression

# We do the estimations for games before 2019
dataset = teams_kpi.search("year < 2019 AND type_tournament = 'UEFA'")
model = LinearRegression("football.team_competition_lr", 
                         solver = 'BFGS')
model.fit(dataset, predictors, response)

display(model)
display(model.regression_report())
display(model.features_importance())
```

### Ranking Prediction

```python
# We compute our predictions
model.predict(teams_kpi, 
              name = "ranking_estimation")

# We smooth our predictions
teams_kpi.analytic('row_number', 
                   order_by = {"ranking_estimation": "asc"}, 
                   by = ["type_tournament", "edition"] , 
                   name = "ranking_prediction")

# Predictions for EURO 2021
teams_kpi.search(conditions = ["""year = 2021 AND team IN ('Belgium', 'Italy', 'Russia', 
                                                           'Poland', 'Ukraine', 'Spain', 
                                                           'France', 'Turkey', 'England', 
                                                           'Czech Republic', 'Finland', 
                                                           'Sweden', 'Croatia', 'Austria', 
                                                           'Netherlands', 'Germany', 'Portugal', 
                                                           'Switzerland', 'Denmark', 'Wales',
                                                           'North Macedonia', 'Hungary', 
                                                           'Slovakia', 'Scotland')
                               AND edition = 15
                               AND type_tournament = 'UEFA'"""],).eval(
                               "ranking_2021",
                               "ROW_NUMBER() OVER(ORDER BY ranking_prediction)")[["team",
                                                                                  "ranking_2021",]].sort("ranking_2021")
```

```python
import shap
import numpy as np

explainer = model.shapExplainer()

# Initialize the JS
shap.initjs()

values = teams_kpi.search("year = 2021 AND edition = 15").sort("ranking_prediction")[predictors][0]

# Getting shap Values
shap_values = explainer.shap_values(np.array([values]))

# Visualize the first prediction's explanation [Germany]
shap.force_plot(explainer.expected_value, 
                shap_values[0,:], 
                values,
                feature_names = predictors,
                out_names = response)
```
